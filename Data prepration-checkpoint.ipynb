# Data prepration and Analysis

Importing Libraries

```python
import requests
import pandas as pd
import numpy as np
```

Downloading data from url


```python
wikipedia_link="https://www.free-4u.com/Colleges/New-York-Colleges.html"
wikipedia_page=requests.get(wikipedia_link).text
```

Web scraping using BeautifulSoup

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(wikipedia_page,'lxml')
print(soup.prettify())
```

Extracting Table from the page

```python
table = soup.find('table')
print(table)
```

```python
Schoollist      = []
Citylist      = []
Enrolllist = []
InStateCostlist=[]
OutStateCostlist=[]
```

```python
for row in table.find_all('tr'):
    cells = row.find_all('td')
    if(len(cells) > 0):
        Schoollist.append(cells[0].text)
        Citylist.append(cells[1].text)
        Enrolllist.append(cells[2].text)
        InStateCostlist.append(cells[3].text)
        OutStateCostlist.append(cells[4].text.rstrip('\n'))
```

```python
ny_school = [('School', Schoollist),
                      ('City', Citylist),
                      ('Enroll', Enrolllist),
                      ('InStateCost',InStateCostlist),
                      ('OutStateCost', OutStateCostlist)]

df = pd.DataFrame.from_dict(dict(ny_school))
df.head()
```

Shape of the dataset

```python
df.shape
```

Handling missing data

```python
df.replace("-", np.nan, inplace = True)

df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
df['InStateCost'].value_counts().idxmax()
```

```python
df["InStateCost"].replace(np.nan, "$18,800.00", inplace=True)
df.head()
```

```python
df['OutStateCost'].value_counts().idxmax()
```

```python
df["OutStateCost"].replace(np.nan, "$19,100.00", inplace=True)
df.head()
```

```python
df['InStateCost'] = df['InStateCost'].map(lambda x:''.join(y for y in x if y.isdigit()) )
df['OutStateCost'] = df['OutStateCost'].map(lambda x:''.join(y for y in x if y.isdigit()))
df['Enroll'] = df['Enroll'].map(lambda x:''.join(y for y in x if y.isdigit()))


df.head()
```

```python
df["InStateCost"] = df.InStateCost.astype(float)
df["OutStateCost"] = df.OutStateCost.astype(float)
df["Enroll"] = df.Enroll.astype(int)
df.dtypes
```

```python
df["OutStateCost"]=(df["OutStateCost"]/100)
df["InStateCost"]=(df["InStateCost"]/100)
df.head()
```

Getting Graduate admisson database 

```python
#grade=pd.read_csv('Admission_Predict.csv')
grade = pd.read_csv('Admission_Predict.csv', sep='\s*,\s*', header=0, encoding='ascii', engine='python')
```

```python
grade.head()
#print(grd.columns.tolist())
Gre_list  = grade['GRE Score'].tolist()
TOFL_list = grade['TOEFL Score'].tolist()
Rat_list  = grade['University Rating'].tolist()
SOP_list  = grade['SOP'].tolist()
Lor_list   = grade['LOR'].tolist()
CGP_list   = grade['CGPA'].tolist()
Res_list  = grade['Research'].tolist()
Chan_list = grade['Chance of Admit'].tolist()
```

```python
df['Gre_score']=Gre_list
df['TOEFL Score']=TOFL_list
df['University Rating']=Rat_list
df['SOP']=SOP_list
df['LOR']=Lor_list
df['CGPA']=CGP_list
df['Research']=Res_list
df['Chance of Admit']=Chan_list
```
